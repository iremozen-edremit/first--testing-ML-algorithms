```python
df = pd.read_excel("Comapany1(DF).xlsx")
df.head()
total = len(df.index)
```

```python
df['new_date'] = [d.date() for d in df['Date/time occurrence']]
df['new_time'] = [d.time() for d in df['Date/time occurrence']]

df.sort_values(by='Date/time occurrence')
```

```python
def f(dt):
    return (7, 59) <= (dt.hour, dt.minute) < (17, 1)

i=0
work=0
for i in range(len(df)):
    if f(df['new_time'][i]) == True:
        work = work +1
```

```python
#Work hours and other hours attacks

total = len(df.index)
other = total - work

labels = ['Work Hours(07:59-17:01)', 'Other Hours(17:01-06:59)']
values = [ work,other]

trace = go.Pie(labels=labels, values=values)
py.iplot([trace], filename='basic_pie_chart')
```
