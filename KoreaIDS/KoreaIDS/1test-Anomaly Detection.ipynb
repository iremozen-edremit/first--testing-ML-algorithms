```python
df2 = df[ (df['Type of malware'] == 'email attachment')]
df2
df2.groupby(['Type of malware','Respnse']).count()
#df['freq'] = df.groupby('new_date')['new_date'].transform('count')
#df2['freq2']= df2.groupby('Respnse')['Respnse'].transform('count')
```

```python
df3 = df2.groupby(['new_date','Respnse']).size().sort_values(ascending=False) \
  .reset_index(name='Sum')

#deleted  == df4
df4 =df3[(df3['Respnse'] == 'deleted')].sort_values(by='new_date')
#none == df5
df5= df3[(df3['Respnse'] == 'none')].sort_values(by='new_date')
#recovered == df6
df6 = df3[(df3['Respnse'] == 'recovered')].sort_values(by=('new_date'))
```

```python
trace = go.Scatter(
    x = df4['new_date'],
    y = df4['Sum']
)

data = [trace]

py.iplot(data, filename='basic-line')
```
